```python
import matplotlib.pyplot as plt

time = [.5,1,1.5, 2, 2.5]
speed = [210,215,225,220,235]

plt.plot(time, speed, color='blue', linestyle='--', linewidth=2, marker='s', label='speed/time')
plt.title('Speed-Time F1 Racing Analysis')
plt.xlabel("Time")
plt.ylabel("Speed")
plt.legend(loc='upper left', fontsize=6)
plt.grid(color='green', linewidth=.5, linestyle=':')
plt.xlim(0,3)
plt.ylim(200, 250)
# plt.xticks([.5,1,1.5, 2, 2.5], ['T1', 'T2', 'T3', 'T4', 'T5'])
# plt.yticks([210,215,225,220,235], ['S1', 'S2', 'S3', 'S4', 'S5'])
plt.show()
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt


days = list(range(1, 16))
traffic = []

for num in range(1, 16):
  randomValue = random.randint(100, 400)
  traffic.append(randomValue)

plt.plot(days, traffic, color='purple', linewidth=2, linestyle='-', marker='o', markerfacecolor='yellow', markeredgewidth=2, markeredgecolor='red', dash_capstyle='round', label='traffic')
# plt.plot(days, traffic, 'bx-.')

fontDictionary = {
    'fontsize':20,
    'fontweight': 'light',
    'color':'green',
    'style':'oblique',
    'family':'serif'
}

plt.xlabel('Days', fontsize=20, fontweight='light', color='green', style='oblique', family='serif',  loc='center', rotation=45, alpha=.8)
plt.ylabel('Web Traffic', fontdict=fontDictionary, labelpad=80, loc='center', rotation=0, alpha=0.5, backgroundcolor='magenta')
plt.show()
```
