```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [1000,5500,6500,2300,4123,9999]

plt.plot(x, y, color='red', linestyle='--', linewidth=2, marker='s', label='month wise sales analysis')
plt.title("Half-Yearly Sales Analysis")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.grid()
plt.legend(loc='upper left', fontsize=20)

plt.show()
```
