# ***Subplots In Matplotlib***

```python
import matplotlib.pyplot as plt
import numpy as np

matches = ['Match1', 'Match2', 'Match3', 'Match4', 'Match5']
virat_score = [45, 78, 66, 89, 101]
rohit_score = [60, 50, 72, 95, 56]

overs = [10,20,30,40,50]
team_score_progress = [160, 188, 247, 358, 256]

players = ['playerA', 'playerB', 'playerC', 'playerD']
goals = [1, 0, 2, 1]


# subplots(nrows, ncols)

fig, ax = plt.subplots(2, 2, figsize=(10, 8))

ax[0,0].plot(matches, virat_score, color='b', marker='o', label="Virat's score")
ax[0,0].plot(matches, rohit_score, color='g', marker='s', linestyle='--', label="Rohit's score")
ax[0,0].legend()

ax[0,1].plot(overs, team_score_progress, color='purple')
ax[0,1].bar(overs, team_score_progress, color='yellow')

ax[1,0].bar(players, goals, color='orange')
ax[1,0].plot(players, goals, c='black', marker='d')


plt.show()
fig.savefig("pathName_given_by_the_user/fileName.png")
```

```python
# DPI
revenue = [50000, 15000, 65000, 60000]
dashboard, plots = plt.subplots(figsize=(6, 4), dpi=300)

plots.plot(revenue)
plots.set_title("Revenue Line Graph")
dashboard.savefig("revenue.jpg")
plt.show()
```

```python
#
days = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri']
temp = [30, 32, 35, 29, 28]
humidity = [60, 65, 70, 68, 64]

fig, ax = plt.subplots(2, 1, sharex=True)

ax[0].plot(days, temp)
ax[0].set_title("Tempreature")
ax[1].plot(days, humidity)
ax[1].set_title("Humidity")

```

```python
store_A = [200, 250, 300, 350]
store_B = [220, 240, 315, 330]

fig, ax = plt.subplots(1, 2, sharey=True)

ax[0].bar(['P1', 'P2', 'P3', 'P4'], store_A)
ax[0].set_title("Store A")
ax[1].bar(['P1', 'P2', 'P3', 'P4'], store_B)
ax[1].set_title("Store B")

plt.show()
```

# **Use of suptitle**

```python
store_A = [200, 250, 300, 350]
store_B = [220, 240, 315, 330]

fig, ax = plt.subplots(1, 2, sharey=True)

ax[0].bar(['P1', 'P2', 'P3', 'P4'], store_A)
ax[0].set_title("Store A")
ax[1].bar(['P1', 'P2', 'P3', 'P4'], store_B)
ax[1].set_title("Store B")

fig.suptitle("Product Prices of Two Stores")

plt.show()
```

# **Use of fig.subplots_adjust**

```python
fig, ax = plt.subplots(2,2, figsize=(12,6))

fig.subplots_adjust(hspace=.5, wspace=.39)

plt.show()
```

# **Use of delaxes**

```python
fig, ax = plt.subplots(2,2, figsize=(12,6))

fig.delaxes(ax[0,0])  #  delete axes
fig.delaxes(ax[1,1])  #  delete axes

plt.show()
```

# **Use of fig.supxlabel & fig.supylabel**

```python
fig, ax = plt.subplots(2,1)

ax[0].plot([1,2,3,4])
ax[1].plot([5,6,7,8])

fig.supxlabel("Time")
fig.supylabel("Performance")

plt.show()
```
