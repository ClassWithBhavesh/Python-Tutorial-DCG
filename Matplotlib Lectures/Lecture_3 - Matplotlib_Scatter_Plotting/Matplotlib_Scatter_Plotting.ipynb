# Bar Graphs in **Matplotlib**

```python
months = ['Jan', 'Feb', 'Mar', 'April', 'May', 'June']
sales = [200, 1500, 700, 350, 1700, 2500]

plt.bar(months, sales, width=.5, linewidth=5, edgecolor='red', color=['purple', 'blue', 'green', 'yellow', 'orange', 'magenta'])
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

```python
# Stacked Bar

maleAudi = [40, 45, 87]
femaleAudi = [33, 65, 74]
slots = ['I', 'II', "III"]

plt.bar(slots, maleAudi, label='Male', zorder=-1)
plt.bar(slots, femaleAudi, label='Female', zorder=2)

plt.legend()
plt.show()
```

```python
# side by side bar graph -
import numpy as np

x = np.arange(len(slots))   # [0,1,2]
width = .2

plt.bar(x - width/2, maleAudi, width, label="Male Audience", color='skyblue', edgecolor='black')
plt.bar(x + width/2, femaleAudi, width, label="Female Audience", color='green', edgecolor='black')

plt.xticks(x, slots)

plt.xlabel("Slots")
plt.ylabel("Audience")
plt.title("Male VS Female Slot Comparison")
plt.legend()
```

# ***Horizontal Bar Graph in Matplotlib***

```python
products = ['Laptop', 'Mobile', 'Camera', 'Mob Stand']

store_A = [60, 35, 122, 2.5]
store_B = [70, 45, 145, 2]

y = np.arange(len(products))  # [0,1,2,3]
height=.35

plt.barh(y - height/2, store_A, height=height)
plt.barh(y + height/2, store_B, height=height)
```
