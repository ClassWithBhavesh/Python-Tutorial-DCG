# ***Pie Chart In Matplotlib***

```python
import matplotlib.pyplot as plt

budget = [5000000, 3000000, 1500000, 7800000, 4200000]

plt.pie(budget)
plt.title("Budget Distribution")
plt.show()
```

```python
plt.pie(budget, labels=["Armed Forces", "MRTD", "Railways", "MHA","Agriculture"])
plt.title("Budget Distrbution")
plt.show()
```

```python
plt.pie(budget, labels=["Armed Forces", "MRTD", "Railways", "MHA","Agriculture"], explode=[0.05, 0.1, 0.3, 0.1, 0.1])
plt.title("Budget Distrbution")
plt.show()
```

```python
plt.pie(budget, labels=["Armed Forces", "MRTD", "Railways", "MHA","Agriculture"], explode=[0.05, 0.1, 0.3, 0.1, 0.1], colors=['green', 'brown', 'yellow', 'skyblue', 'magenta'])
plt.title("Budget Distrbution")
plt.show()
```

```python
plt.pie(budget, labels=["Armed Forces", "MRTD", "Railways", "MHA","Agriculture"], colors=['green', 'brown', 'yellow', 'skyblue', 'magenta'], autopct='%1.0f%%')
plt.title("Budget Distrbution")
plt.show()
```

```python
plt.pie(budget, labels=["Armed Forces", "MRTD", "Railways", "MHA","Agriculture"], autopct='%1.0f%%', pctdistance=.8)
plt.title("Budget Distrbution")
plt.show()
```

```python
plt.pie(budget, labels=["Armed Forces", "MRTD", "Railways", "MHA","Agriculture"], autopct='%1.0f%%', pctdistance=.8, labeldistance=1.3)
plt.title("Budget Distrbution")
plt.show()
```

```python
plt.pie(budget, labels=["Armed Forces", "MRTD", "Railways", "MHA","Agriculture"], autopct='%1.0f%%', pctdistance=.8, labeldistance=1.3, startangle=180)
plt.title("Budget Distrbution")
plt.show()
```

```python
plt.pie(budget, labels=["Armed Forces", "MRTD", "Railways", "MHA","Agriculture"], explode=[.1, 0, 0, 0.1, 0], autopct='%1.0f%%', pctdistance=.8, labeldistance=1.3, startangle=180, shadow=True)
plt.title("Budget Distrbution")
plt.show()
```

```python
plt.pie(budget, labels=["Armed Forces", "MRTD", "Railways", "MHA","Agriculture"], autopct='%1.0f%%', startangle=180, wedgeprops={'edgecolor': 'black', 'linewidth': .5, 'width': .7})
plt.title("Budget Distrbution")
plt.show()
```

```python
plt.pie(budget, labels=["Armed Forces", "MRTD", "Railways", "MHA","Agriculture"], autopct='%1.0f%%', startangle=180, textprops={'fontsize': '12', 'color': 'red', 'backgroundcolor': 'yellow'})
plt.title("Budget Distrbution")
plt.show()
```

```python
plt.pie(budget, labels=["Armed Forces", "MRTD", "Railways", "MHA","Agriculture"], autopct='%1.0f%%', startangle=180, counterclock=False)
plt.title("Budget Distrbution")
plt.show()
```

```python
plt.pie(budget, labels=["Armed Forces", "MRTD", "Railways", "MHA","Agriculture"], autopct='%1.0f%%', startangle=180, counterclock=False, radius=3.3)
plt.title("Budget Distrbution")
plt.show()
```

```python
plt.pie(budget, labels=["Armed Forces", "MRTD", "Railways", "MHA","Agriculture"], autopct='%1.0f%%', startangle=180, counterclock=False, radius=3.3, frame=True)
plt.title("Budget Distrbution")
plt.show()
```
