```python
import pandas as pd

data = {
    "Month": ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"],
    "Sales": [1200,1500,1600,1800,2100,2500,2400,2300,2200,2000,1900,1700],
    "Customers": [300,350,400,420,500,550,530,520,510,480,460,430],
    "Returns": [30,25,40,35,20,15,25,20,30,35,40,45]
}
df = pd.DataFrame(data).set_index("Month")
df
```

```python
df.plot()  # by default ---> kind='line'
```

```python
df.plot(kind='line')
```

```python
df.plot(kind='bar')  # by default ---> stacked="False"
```

```python
df.plot(kind='bar', stacked=True)
```

```python
df.plot(kind='box')
```

```python
df.plot(kind='hist')
```

```python
df.plot(kind='line', xlabel='Months', ylabel='Sales')
```

```python
df.plot(kind='barh')
```

```python
df.plot(kind='barh', stacked=True)
```
